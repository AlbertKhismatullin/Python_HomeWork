1. Вывести график функции y = x**2 + 3x - 1
2. Описать названия осей и графика
3. Увеличить масштаб фигуры по своему усмотрению
4. На этой же фигуре сделать график y = sin(x) * 3
5. Сделать график (4) красной прерывистой линией


```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-8, 8, 100)
plt.title('Графики функций |y| = x^2 + 3x - 1, y1 = sin(x)*3')
plt.ylabel('Ось y')
plt.xlabel('Ось x')
plt.grid()
plt.plot(x, x ** 2 + 3 * x - 1)
plt.plot(x, np.sin(x) * 3, 'r--')
plt.figure(figsize=(10, 6))
plt.show()
```

Вывести диаграммы количества алмазов:

1.различного состояния
2.различной прозрачности

Каждую из диаграмм выведите в интерфейсе subplot в n горизонтально расположенных окошках, где n - количество состояний и видов прозрачности



```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv ('diamonds.csv')
cuts = df['cut'].value_counts()
plt.subplot(3, 3, 1)
plt.bar(x=cuts.index, height=cuts.values)
plt.title('Кол-во алмазов в зависимости от состояния огранки')
plt.ylabel('Кол-во алмазов')
plt.xlabel('Огранка')
plt.grid()
colors = df['color'].value_counts()
plt.subplot(3, 3, 6)
plt.bar(x=colors.index, height=colors.values)
plt.title('Кол-во алмазов в зависимости от цвета')
plt.ylabel('Кол-во алмазов')
plt.xlabel('Цвет')
plt.grid()
plt.show()
```

Для алмазов состояния Ideal вывести график зависимости массы камня (в каратах) от цены (будет более целесообразно сделать график из точек, т.е. с дискретными значениями). Есть ли основания полагать, что для данного состояния характерна прямая зависимость цены от массы кристалла?

Ответ: зависимость цены от массы существует

```python
df = pd.read_csv ('diamonds.csv')
df = df[df['cut'] == 'Ideal']
plt.scatter(x=df['carat'], y=df['price'])
plt.title('Зависимость массы алмазов от цены')
plt.xlabel('масса')
plt.ylabel('цена')
plt.grid()
plt.show()
```
